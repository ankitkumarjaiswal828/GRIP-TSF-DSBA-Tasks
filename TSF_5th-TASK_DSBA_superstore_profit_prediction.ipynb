# Visualized increasing and decreasing  values of profit through lineplot graph

```python
plt.figure(figsize = (15,5))
sns.lineplot(x = 'Discount', y=  'Profit',label = 'Profit' ,color = 'r', data = df )
plt.legend()
plt.show()
```

```python
#visualized the profits and sales across different segments
df.groupby('Segment')[['Profit','Sales'] ].sum().plot.bar(color = ['r', 'b'],figsize = (15,5),fontsize = 15)
plt.ylabel('Profit and Sales',fontsize = 15)
plt.show()
```

```python
#visualize the segments wise sales distribution in each region
plt.figure(figsize = (15,8))
plt.title('Distribution of segment wise in each region',fontsize = 15)
sns.barplot(x = 'Region', y = 'Sales' , data = df , hue = 'Segment',order = df['Region'].value_counts().index, palette = 'Greens')
plt.xlabel('Region' ,fontsize =15)
plt.ylabel('Sales',fontsize = 15)
plt.tick_params(labelsize = 15)
plt.show()

```

```python
#visualize the profit and sales in defferent region
df.groupby('Region')[['Profit' , 'Sales']].sum().plot.bar(color = ['r','g'],figsize = (15,5))
plt.ylabel('Profits and Sales',fontsize = 15)
plt.show()
```

```python
#visualize the profit and sales across Satetes
st = df.groupby('State')[['Sales','Profit']].sum().sort_values(by = 'Sales',ascending = False)
st[:].plot.bar(color = ['b','r'],figsize = (15,5))
plt.xlabel('States',fontsize = 15)
plt.ylabel('Profit and sales', fontsize = 15)
plt.show()
```

```python
#visualize the sales and profit across different categories
df.groupby('Category')[['Sales','Profit']].sum().plot.bar(color = ['b','r'],figsize = (15,5))
plt.ylabel('Sales and profit' , fontsize = 15)
plt.xlabel("Category",fontsize = 15)
plt.show()
```

```python
#visualize the profit and sales across the different Sub-categories
sub_ca = df.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by = 'Sales' , ascending = False)
sub_ca.plot.bar(color = ['g', 'r'], figsize = (15,5))
plt.title('Visualizing the profit and Sales Across the sub-categories' , fontsize  = 15)
plt.xlabel('Sub-Categories' , fontsize =15)
plt.ylabel('Profit and Sales',fontsize = 15)
plt.show()
```
